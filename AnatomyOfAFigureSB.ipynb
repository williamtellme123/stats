```python
# This figure shows the name of several matplotlib elements composing a figure
# The following config code is optional
# It allows intellisense to work more quickly using the tab key
%config Completer.use_jedi=False
%config IPCompleter.greedy=True
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator

np.random.seed(19680801)
# import matplotlib.transforms

X = np.linspace(0.5, 3.5, 100)
Y1 = 3 + np.cos(X)
Y2 = 1 + np.cos(1 + X / 0.75) / 2
Y3 = np.random.uniform(Y1, Y2, len(X))
#  -----------------------------------------------------

def minor_tick(x, pos):
    if not x % 1.0:
        return ""
    return f"{x:.2f}"

#  -----------------------------------------------------
fig = plt.figure(figsize=(8, 8))  # creates
plt.xlim(0, 4)
plt.minorticks_on()
plt.xticks(np.arange(5))
plt.yticks(np.arange(5))
plt.tick_params(which="major", width=1.0, length=10)
plt.tick_params(which="minor", width=1.0, length=5, labelsize=10, labelbottom=True, labelcolor="C0")

plt.gca().xaxis.set_minor_locator(AutoMinorLocator(4))
plt.gca().xaxis.set_minor_formatter(minor_tick)


plt.title("Anatomy of a Pyplot Figure", fontsize=20, verticalalignment="bottom")
plt.plot(X, Y1, c=(0.25, 0.25, 1.00), lw=2, label="Blue signal", zorder=10)
plt.plot(X, Y2, c=(1.00, 0.25, 0.25), lw=2, label="Red signal")
plt.plot(X, Y3, linewidth=0, marker="o", markerfacecolor="w", markeredgecolor="k")
plt.xlabel("X axis label")
plt.ylabel("Y axis label")
plt.legend()
plt.grid()
plt.tight_layout()

def circle(x, y, radius=0.03):
    from matplotlib.patches import Circle
    circle = Circle(
        (x, y),
        radius,
        clip_on=False,
        zorder=10,
        linewidth=1,
        edgecolor="black",
        facecolor=(0, 0, 0, 0.0125)
      )
    fig.add_artist(circle)

def text(x, y, text):
    plt.text(
        x,
        y,
        text,
        backgroundcolor="white",
        ha="center",
        va="top",
        weight="bold",
        color="blue",
    )


plt.annotate(
    "",
    xy=(3.15, 0.33),
    xytext=(3.4, 0.75),
    arrowprops=dict(arrowstyle='->',connectionstyle="arc3",color="blue"),
    size=12, 
    color="blue",
    weight="bold"
)
plt.annotate(
    "Spines",
    xy=(3.98, 0.6),
    xytext=(3.4, 0.75),
    arrowprops=dict(arrowstyle='->',connectionstyle="arc3",color="blue"),
    size=12, 
    color="blue",
    weight="bold")

plt.text(3.69, -0.08,
        "Made with https://matplotlib.org",
        size=10, 
        ha="right",
        color=".5")

# Minor Tick Label
circle(.19, 0.07)
text(0.50, 0.08, "Minor tick label")

# Major tick
circle(.080, .93)
text(0.03, 3.83, "Major tick")

# Major tick label
circle(.05, .70)
text(-0.10, 2.85, "Major tick label")

# X Label
circle(.49, 0.03)
text(1.85, -0.08, "X axis label")

# Y Label
circle(.025, .48)
text(-0.29, 1.76, "Y axis label")

# Title
circle(.399, .97)
text(1.43, 3.97, "Title")

# Blue plot
circle(.464, .65)
text(1.73, 2.60, "Line\n(line plot)")

# # Red plot
# circle(1.20, 0.60)
# text(1.20, 0.40, "Line\n(line plot)")

# Scatter plot
circle(.80, .40)
text(3.23, 1.55, "Markers\n(scatter plot)")

# Grid
circle(.752, .698)
text(3.00, 2.83, "Grid")

# Legend
circle(.918, .898)
text(3.72, 3.69, "Legend")

# Axes
circle(0.16, 0.17)
text(0.370, 0.54, "Axes")

# Figure
# circle(-0.3, 0.65)
circle(.025, .18)
text(-0.24, 0.59, "Figure")


# Red plot
circle(0.35, 0.14)
text(1.215, 0.39, "Line\n(line plot)")
```
